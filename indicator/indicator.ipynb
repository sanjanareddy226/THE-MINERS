```python
import cv2
argv="/Users/sanjanasrinivasareddy/Downloads/z.mov"
vid = cv2.VideoCapture(argv) 
```

```python
lastNorm = 0.0;
lastCounter = 0;
counter = 0;
currentState = 0;
i=0
```

```python
import time
timeout = time.time() + 7   # 5 minutes from now
while(True): 
    test = 0
    if test == 5 or time.time() > timeout:
        break
    test = test - 1
    ret, frame = vid.read() 
    currentNorm = cv2.norm(frame)
    diffNorm = currentNorm - lastNorm
    if (diffNorm > 0 and currentState == 0):
        currentState = 1
        print("light was on for   ")
        print(counter - lastCounter)
        lastCounter = counter
    if (diffNorm < 0 and currentState == 1):
        
        currentState = 0
        print("light was off for  ")
        print(counter - lastCounter)
        lastCounter = counter
        
    lastNorm = currentNorm
    counter=counter+1
```
